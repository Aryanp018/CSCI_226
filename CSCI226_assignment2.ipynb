```python
# --- Exercise 6.2.3 ---

# a) Ships heavier than 35,000 tons
cur.execute("""
    SELECT class, displacement
    FROM Classes
    WHERE displacement > 35000;
""")
rows = cur.fetchall()
df2_a = pd.DataFrame(rows, columns=["class", "displacement"])
print("Exercise 6.2.3 - Query a Results:")
display(df2_a)

# b) Battle of Guadalcanal
cur.execute("""
    SELECT name, date
    FROM Battles
    WHERE name = 'Guadalcanal';
""")
rows = cur.fetchall()
df2_b = pd.DataFrame(rows, columns=["battle_name", "date"])
print("Exercise 6.2.3 - Query b Results:")
display(df2_b)

# c) All entries (classes + battles)
cur.execute("""
    SELECT class AS name FROM Classes
    UNION
    SELECT name FROM Battles;
""")
rows = cur.fetchall()
df2_c = pd.DataFrame(rows, columns=["name"])
print("Exercise 6.2.3 - Query c Results:")
display(df2_c)
```
