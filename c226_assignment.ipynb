```python
# Simplified query cell (uses session-level search_path)
# - Connect with the STUDENT URL
# - Set search_path so we can omit schema prefixes
# - Run example queries and show results

import os, getpass, pandas as pd
from sqlalchemy import create_engine, text

STUDENT_DB_URL = os.environ.get("NEON_STUDENT_DB_URL") or getpass.getpass("Paste STUDENT DB URL (hidden): ")
engine = create_engine(STUDENT_DB_URL, pool_pre_ping=True, future=True)

# -------------------------------
# Query 1: List all products (maker + model) from product table
# -------------------------------
with engine.connect() as conn:
    q1_sql = """
    SELECT maker, model, type
    FROM product
    LIMIT 20;
    """
    q1_df = pd.read_sql_query(text(q1_sql), conn)

print("Query 1 Results: Products (maker, model, type)")
display(q1_df)

# -------------------------------
# Query 2: Top 5 most expensive laptops
# -------------------------------
with engine.connect() as conn:
    q2_sql = """
    SELECT p.maker, l.model, l.price
    FROM product p
    JOIN laptop l USING (model)
    ORDER BY l.price DESC
    LIMIT 5;
    """
    q2_df = pd.read_sql_query(text(q2_sql), conn)

print("\nQuery 2 Results: Top 5 Most Expensive Laptops")
display(q2_df)

# -------------------------------
# Query 3: Count PCs by speed (GHz)
# -------------------------------
with engine.connect() as conn:
    q3_sql = """
    SELECT speed, COUNT(*) AS num_pcs
    FROM pc
    GROUP BY speed
    ORDER BY speed DESC;
    """
    q3_df = pd.read_sql_query(text(q3_sql), conn)

print("\nQuery 3 Results: Count of PCs by Speed")
display(q3_df)
```
